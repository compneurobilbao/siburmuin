```python
import os

def anonymize_images(root_dir):
    """
    Recorrer todas las carpetas en la ruta raz y cambiar el nombre de los archivos de imgenes
    para anonimizar el nombre, usando el nombre de la carpeta como nuevo nombre de archivo.
    
    Args:
        root_dir (str): La ruta raz donde se encuentran las carpetas con las imgenes.
    """
    if not os.path.exists(root_dir):
        print("La ruta especificada no existe.")
        return

    for foldername, subfolders, filenames in os.walk(root_dir):
        
        folder_name = os.path.basename(foldername)
        
        for filename in filenames:
            
            name, ext = os.path.splitext(filename) # Separar el nombre del archivo y la extension

            if ext.lower() in ['.jpg', '.jpeg', '.png', '.gif']:
                if 'OS' in name or 'OD' in name:
                    # Encontrar el ndice de 'OS' o 'OL'
                    if 'OS' in name:
                        index = name.index(' OS ')
                    else:
                        index = name.index(' OD ')

                    
                    # Tomar solo la parte despus de 'OS' o 'OL' y quitar espacios en blanco
                    clean_name = name[index + 2:].strip()  # +2 para omitir 'OS' o 'OL'
                    new_filename = f"{folder_name} {clean_name}{ext}"

                    old_file_path = os.path.join(foldername, filename)
                    new_file_path = os.path.join(foldername, new_filename)

                    # Renombrar el archivo
                    os.rename(old_file_path, new_file_path)
                    print(f"Renombrado: {old_file_path} a {new_file_path}")
                else:
                    print(f"PRECAUCION!!! - Saltando imagen que no sigue la estructura: {filename}")


root_directory = r'D:\BEGIALZ_BASELINE'

anonymize_images(root_directory)
```
