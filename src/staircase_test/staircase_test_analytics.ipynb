```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
data = pd.read_csv('./data/trials_data.csv')

# Filtrar solo las filas donde ocurrieron reversals
reversal_data = data[data['reversals'] > 0]

# Calcular el umbral de contraste como la media de las ltimas n inversiones (reversals)
n_reversals_to_average = 9  # En general, se toma la media de las ltimas n inversiones
threshold = reversal_data['contrast'].tail(n_reversals_to_average).mean()

# Graficar el contraste del estmulo a lo largo de los ensayos
plt.figure(figsize=(10, 6))
plt.plot(data['trial'], data['contrast'], marker='o', linestyle='-', label='Stimulus Contrast')

# Aadir una lnea horizontal para el umbral de contraste
plt.axhline(y=threshold, color='r', linestyle='--', label=f'Threshold = {threshold:.3f}')

# Etiquetas y ttulo
plt.xlabel('Number of Trials')
plt.ylabel('Stimulus Contrast')
plt.title('Stimulus Contrast Across Trials')
plt.legend()
plt.grid(True)

# Mostrar la grfica
plt.show()

# Mostrar el valor del umbral calculado
print(f"Umbral de contraste estimado: {threshold:.3f}")
```
